```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load the movie rating data into a dataframe
rating_header = ["user_id", "item_id", "rating", "timestamp"]
rating = pd.read_csv("u.data", sep = '\t', header = None, names=rating_header)
rating.head()
```

```python
# occupation 職業
user_header = ["user_id", "age", "gender", "occupation", "zip_code"]
users = pd.read_csv("u.user", sep = '|', header = None, names=user_header)
users.head()
```

```python
movie_header = ["item_id", "title", "release_date", "video_release_date", "IMDb_URL",
         "unknown", "Action", "Adventure", "Animation","Children's", "Comedy", "Crime",
         "Documentary", "Drama", "Fantasy", "Film-Noir", "Horror", "Musical", "Mystery", 
         "Romance", "Sci-Fi", "Thriller", "War", "Western"]
movies = pd.read_csv("u.item", sep = '|', header = None, encoding = 'latin1', names = movie_header)
movies.head()
```

### 數值轉化

```python

plt.boxplot(rating['rating'], labels=['rating'])
plt.title('Movie Rating Box Plot')
plt.xlabel('Column Name')
plt.ylabel('Rating')
plt.show()
```

```python
movie_genres = movie_header[-18:]

genres_count_list = []

for genre in movie_genres:
    genres_count_list.append(sum(movies[genre]))

print(movie_genres)
print(genres_count_list)
                             
genre_df = pd.DataFrame({'count':genres_count_list, 'genres': movie_genres})

genre_df.head()

plt.figure(figsize=(20, 15))
plt.bar(genre_df['genres'], genre_df['count'],  width=0.5)
plt.show()
```
